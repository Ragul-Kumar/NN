```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.metrics import classification_report 
from sklearn.svm import SVC 
from sklearn.model_selection import train_test_split 

# Set random seed for reproducibility
np.random.seed(0) 

# Generate random data
X = np.random.randn(200, 2) 
print("X:\n", X[:5]) 

# Create XOR target labels
y_xor = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0).astype(int) 
print("output",y_xor)

# Plotting function
def plot_gate(X, y, title): 
    plt.title(title) 
    plt.scatter(X[y == 1, 0], X[y == 1, 1], c='b', marker='x', label='1') 
    plt.scatter(X[y == 0, 0], X[y == 0, 1], c='r', marker='o', label='0') 
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show() 

# Visualize XOR data
plot_gate(X, y_xor, "XOR Data Distribution") 

# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(X, y_xor, test_size=0.25, random_state=42) 

# Train an SVM model with polynomial kernel
model = SVC(kernel='poly', degree=2, coef0=1) 
model.fit(train_data, train_labels) 

# Print classification report
print("Classification Report:\n") 
print(classification_report(test_labels, model.predict(test_data))) 
```
